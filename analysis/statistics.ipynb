```python
import glob
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import PandasTools

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_to_library = '../FragmentLibrary'

# list of folders for each subpocket
folders = glob.glob(path_to_library+'/*')
subpockets = [folder[-2:] for folder in folders]
subpockets
```

```python
# read data

data = pd.DataFrame()

# iterate over subpockets
for i, folder in enumerate(folders):
    
    subpocket = subpockets[i]
    
    file = folder+'/'+subpocket+'.sdf' 
    df = PandasTools.LoadSDF(file, embedProps=True)
        
    fragments = []
    fingerprints = []
    smiles = []

    for frag in df.ROMol:
        # smiles for drawing fragment
        smile = Chem.MolToSmiles(frag)
        smiles.append(smile)
        # 2D fragment
        fragments.append(Chem.MolFromSmiles(smile))
     
    df['smiles'] = smiles
    df['fragment'] = fragments
    df['subpocket'] = subpocket
    
    data = data.append(df)
    
data = data.reset_index(drop=True)
data.shape
```

```python
def is_single(fragment):
    atom_symbols = [atom.GetSymbol() for atom in fragment.GetAtoms()]
    return '*' not in atom_symbols
mask_single = data['ROMol'].apply(is_single)
```

Number of double and single bonds:

```python
bond_types = {}
for fragment in data.ROMol:
    dummy_atoms = [atom for atom in fragment.GetAtoms() if atom.GetSymbol()=='*']
    for dummy in dummy_atoms:
        atom = dummy.GetNeighbors()[0]
        bond_type = fragment.GetBondBetweenAtoms(dummy.GetIdx(), atom.GetIdx()).GetBondType()
        if bond_type in bond_types:
            bond_types[bond_type] += 1
        else:
            bond_types[bond_type] = 1
bond_types
```
