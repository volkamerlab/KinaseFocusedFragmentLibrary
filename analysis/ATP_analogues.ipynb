```python
from rdkit import Chem
from rdkit.Chem import PandasTools
import pandas as pd
```

```python
df = pd.read_csv('../ATP_analogues.txt', delimiter='\t', header=None, names=['KLIFS', 'Name', 'SMILES'])
PandasTools.AddMoleculeColumnToFrame(df,'SMILES','Molecule')
print('The following ligands were discarded:')
df.drop(columns=['SMILES'])
```

```python
# ligands containing ribose (found by clustering)
df = pd.read_csv('../riboses.txt', delimiter='\t', header=None, names=['KLIFS', 'Name', 'SMILES'])
PandasTools.AddMoleculeColumnToFrame(df,'SMILES','Molecule')
print('The following ligands were included:')
df.drop(columns=['SMILES'])
```
